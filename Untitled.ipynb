```python
#importing modules
import pandas as pd
import numpy as np
import math
import datacompy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading dataframs to analyze and cleaning datasets
sal = pd.read_csv("Salary.csv")
stats = pd.read_csv('TeamWins.csv')
stats.drop(stats.iloc[:, 4:], inplace = True, axis = 1)
stats.dropna(inplace = True)


print(sal)
print(stats)

```

In the previous cell, we have just gotten our data loaded into a dataframe using pandas. From here, we can go on to interpretate the date to see if the winning percentage has a correlation with the millions in pay roll a team is spending. In this cell below, we are comparing the data based on a shared index to see if we are missing any values. 

```python
compare = datacompy.Compare(sal, stats, on_index = True)
print(compare.report())
output = pd.concat([sal, stats], axis=1)
print(output)
```

```python
print(output.keys())
output = output.loc[:,~output.columns.duplicated()]
output = output.iloc[0:508]
```

```python
print(output.keys())
print(output)
```

In the previous cell, we merged togeather the two dataframes and deleted any duplicate entries for the column headers. In the next cell, I will compute some statistcs based off of winning percentage and payroll for each team and the overall league. 

```python
teams = output['team'].unique()
print(teams)
```

```python
teams_map = {}
for i in teams:
    teams_map[i] = output.loc[output['team'] == i]
team_stat = {}

for key, value in teams_map.items():
    avgpr = value['payroll'].mean()
    avgwp = value['winning percentage'].mean()
    team_stat[key] = [avgpr, avgwp]
print("Team, Average Payroll, Winning Percentage")
totalPay = []
totalWin = []
teamList = []
for key, value in team_stat.items():
    teamList.append(key)
    totalPay.append(value[0])
    totalWin.append(value[1])
    print(key, value)
plt.figure(figsize=(20,20))
plt.scatter(totalPay, totalWin)
for i in range(len(totalPay)):
    plt.annotate(teamList[i], (totalPay[i], totalWin[i]))

plt.xlabel("Payroll in Millions")
plt.ylabel("Win Percentage")
plt.title("Payroll in Millions vs Win Percentage",fontsize=15)
AvgPR = sum(totalPay) / len(totalPay)
AvgWin =  sum(totalWin) / len(totalWin)
print(AvgPR, AvgWin)
a, b = np.polyfit(totalPay, totalWin, 1)
newTP = []
for i in range(len(totalPay)):
    newTP.append(totalPay[i] * a)
print("Coeffecient: " + str(a) , " Intercept = " + str(b))
plt.plot(totalPay, newTP+b, color='steelblue', linestyle='--', linewidth=2)
```

In the previous cell, we plotted the correlation of all the teams payroll in millions over the past two decades in comparision to the win percentage they are able to accomplish. Furthermore, by using simple linear regression through a * x + b, I was able to plot the line of best fit. In the following cells, I will be plotting a regression model for each teams future payroll versus wins. 

```python
from sklearn import linear_model
teamModels = {}
counter = 0
for key, value in teams_map.items():
    teamName = key
    dataFrame = value
    X = dataFrame['payroll']
    y = dataFrame['winning percentage']
    dates = dataFrame['year']
    reg = linear_model.LinearRegression()
    reg.fit(pd.DataFrame(X), y)
    co = reg.coef_[0]
    inter = reg.intercept_
    print("" + key + " Coeffecient: " + str(co) , " Intercept = " + str(inter))
    plt.figure(figsize=(10,10))
    plt.scatter(X, y)
    newTP = []
    for i in range(len(dates)):
        plt.annotate(dates[counter], (X[counter], y[counter]))
        newTP.append(co * X[counter])
        counter += 1

    plt.xlabel("Payroll in Millions")
    plt.ylabel("Win Percentage")
    plt.title( "" + teamName + " Payroll in Millions vs Win Percentage",fontsize=15)
    plt.plot(X, newTP+inter, color='steelblue', linestyle='--', linewidth=2)
```
